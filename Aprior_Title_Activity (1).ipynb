```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import ast
```

```python
df = pd.read_excel(r"C:\Users\bmccormack\Downloads\Assignment_Status___ACT_Q1_2025_Report.xlsx")
```

```python
df.head()
```

```python
grouped = df.groupby('Title')['Activity Name'].apply(list).reset_index()
grouped
```

```python
grouped['Activity Name'].value_counts()
```

```python
grouped['Activity Name'] = grouped['Activity Name'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

df_exploded = df.explode('Activity Name')

#encoded dataframe
df_encoded = df_exploded.groupby(['Title', 'Activity Name']).size().unstack(fill_value=0)
df_encoded = df_encoded.astype(bool)

#apriori
frequent_itemsets = apriori(df_encoded, min_support=0.2, use_colnames=True)

#association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.9)

frequent_itemsets.to_csv("frequent_itemsets.csv", index=False)
rules.to_csv("association_rules.csv", index=False)

print("Apriori results saved as 'frequent_itemsets.csv' and 'association_rules.csv'.")
```
