```python
import matplotlib.pyplot as plt
import seaborn as sns

#Sum the SalesInThousands by Promotion
sales_by_promotion = df.groupby('Promotion')['SalesInThousands'].sum().reset_index()

#Custom color palette for promotions (keys as strings)
promotion_colors = {'1': 'blue', '2': 'green', '3': 'orange'}

# Convert Promotion column to string for matching the keys in the palette
sales_by_promotion['Promotion'] = sales_by_promotion['Promotion'].astype(str)

#Barplot with summed sales for each promotion and custom colors
plt.figure(figsize=(8, 6))
sns.barplot(x='Promotion', y='SalesInThousands', data=sales_by_promotion, hue='Promotion', palette=promotion_colors, legend=False)
plt.title('Total Sales by Promotion', fontsize=16)
plt.xlabel('Promotion', fontsize=12)
plt.ylabel('Total Sales (in Thousands)', fontsize=12)
plt.show()
```

```python
from scipy import stats

#Split the data based on the 'Promotion' column (1, 2, 3)
promotion_groups = df['Promotion'].unique()
print(f"Unique Promotion Groups: {promotion_groups}")

# Split the sales data by promotion
group_a = df[df['Promotion'] == promotion_groups[0]]['SalesInThousands']
group_b = df[df['Promotion'] == promotion_groups[1]]['SalesInThousands']
group_c = df[df['Promotion'] == promotion_groups[2]]['SalesInThousands']

#Perform ANOVA to compare means across the three groups
f_stat, p_value = stats.f_oneway(group_a, group_b, group_c)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

# If the p-value is less than 0.05, the result is deemed statistically significant
if p_value < 0.05:
    print("We reject the null hypothesis: There is a significant difference in sales between the three promotion groups.")
else:
    print("We fail to reject the null hypothesis: There is no significant difference in sales between the three promotion groups.")

# Visualization
# Fix the color mappings: 1 -> Blue, 2 -> Green, 3 -> Orange
promotion_colors = {1: 'blue', 2: 'green', 3: 'orange'}

# Correct histogram plot with proper color assignments
plt.figure(figsize=(10, 6))

# Assign correct colors based on the promotion number
for i, promotion in enumerate(promotion_groups):
    if promotion == 1:
        color = promotion_colors[1]
    elif promotion == 2:
        color = promotion_colors[2]
    elif promotion == 3:
        color = promotion_colors[3]

    # Plot each group with the correct color
    group = df[df['Promotion'] == promotion]['SalesInThousands']
    plt.hist(group, alpha=0.7, label=f'Promotion {promotion}', color=color, bins=20)

plt.legend(loc='upper right')
plt.title('Sales Distribution: Different Promotions', fontsize=16)
plt.xlabel('Sales (in Thousands)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
#Magnitude of Impact
means = {promotion_groups[0]: np.mean(group_a), 
         promotion_groups[1]: np.mean(group_b), 
         promotion_groups[2]: np.mean(group_c)}

baseline = min(means, key=means.get)
print(f"Baseline Promotion: {baseline}")

for promo, mean_sales in means.items():
    if promo != baseline:
        lift = (mean_sales - means[baseline]) / means[baseline] * 100
        print(f"Relative lift for Promotion {promo} over Promotion {baseline}: {lift:.2f}%")

#Bayesian Probability of Being Best using Bootstrap Sampling
np.random.seed(42)
def bootstrap_prob_best(groups, n_samples=10000):
    samples = {promo: np.random.choice(group, size=n_samples, replace=True) for promo, group in groups.items()}
    best_counts = {promo: 0 for promo in groups.keys()}
    
    for i in range(n_samples):
        best_promo = max(samples, key=lambda promo: samples[promo][i])
        best_counts[best_promo] += 1
    
    return {promo: count / n_samples for promo, count in best_counts.items()}

#Compute probabilities
probabilities = bootstrap_prob_best({promotion_groups[0]: group_a, 
                                     promotion_groups[1]: group_b, 
                                     promotion_groups[2]: group_c})

for promo, prob in probabilities.items():
    print(f"Probability Promotion {promo} is best: {prob:.2%}")
```

**Comparing the three promotions using Bayesian Statistics and Magnitude Impact, we see that promotion 1 is deemed the best, but why? Promotion 3 had the most sales. This could be a result of Promotion 3 having a higher average sales value and/or lower variance. Let's plot the Sales Distribution by Promotion.**

```python
plt.figure(figsize=(10, 6))

#Fix color mappings
promotion_colors = {1: 'blue', 2: 'green', 3: 'orange'}

#Create twin axis for KDE plots
ax1 = plt.gca()  
ax2 = ax1.twinx() 

#Plot histograms and KDEs
for promotion in promotion_groups:
    color = promotion_colors[promotion]
    group = df[df['Promotion'] == promotion]['SalesInThousands']

    #Histogram
    ax1.hist(group, alpha=0.4, label=f'Promotion {promotion}', color=color, bins=20, edgecolor='black', density=True)

    #KDE plot
    sns.kdeplot(group, color=color, linewidth=2, ax=ax2)

    #Add mean sales as a vertical line
    mean_sales = group.mean()
    ax1.axvline(mean_sales, color=color, linestyle='dashed', linewidth=2, label=f'Mean {promotion}')

#Formatting
ax1.set_xlabel('Sales (in Thousands)', fontsize=12)
ax1.set_ylabel('Frequency', fontsize=12)
ax2.set_ylabel('Density (KDE)', fontsize=12)
plt.title('Sales Distribution by Promotion', fontsize=16)
ax1.legend(loc='upper right')

plt.show()
```

**Promotion 3 (Orange) has the highest mean sales (rightmost dashed line).
Promotion 1 (Blue) has a narrow KDE peak, meaning stable performance.
Promotion 3 (Orange) has a wide KDE curve, meaning high variability.**

**To conclude, Promotion 3 has the highest average sales but is more risky (inconsistent). Promotion 1 is more stable, so it may be the better choice if predictability is important.**
